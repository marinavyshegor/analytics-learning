```python
h = input()
a = float(input())
if h == 'круг':
    p = 3.14
    print(p * (a ** 2))  
elif h == 'прямоугольник':
    b = float(input())
    print (a * b)
elif h == 'треугольник':
    b = float(input())
    c = float(input())
    p = (a + b + c) / 2
    print((p * (p-a) * (p-b) * (p-c)) ** 0.5)
```
