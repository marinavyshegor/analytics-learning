```python
programmers = int(input())
a = programmers % 100
b = programmers % 10
if 11 <= a <= 19:
    print(f"{programmers} программистов")
elif b == 1:
    print(f"{programmers} программист")
elif 2 <= b <= 4:
    print(f"{programmers} программиста")
else:
    print(f"{programmers} программистов")
```

```python
a = int(input())
b = a // 100000 % 10
c = a // 10000 % 10
d = a // 1000 % 10
e = a // 100 % 10
f = a // 10 % 10
g = a % 10
if b + c + d == e + f + g:
    print('Счастливый')
else:
    print('Обычный')
```
