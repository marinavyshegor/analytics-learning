# День 2: Переменные, типы данных, f-strings
## Практика на реальных SEO-данных

```python
sessions = 1240
bounce_rate = 0.54
avg_position = 8.3
```

```python
print(type(sessions))
print(type(bounce_rate))
print(type(avg_position))
```

```python
site_name = "my-shop.ru"
page_url = "/catalog/women-dresses/"
```

```python
print(type(site_name))
print(type(page_url))
```

```python
print(f"Сайт: {site_name}")
print(f"Страница: {page_url}")
print(f"Сессий: {sessions}, отказы: {bounce_rate * 100}%, позиция: {avg_position}")  
```

```python
print(f"Сессий: {sessions}, отказы: {bounce_rate * 100:.0f}%, позиция: {avg_position:.1f}")
```

## Выводы дня

- Переменные — это именованные контейнеры для данных
- Основные типы: int, float, str
- F-strings позволяют вставлять переменные прямо в текст
- Форматирование :.0f и :.1f делает вывод чистым
