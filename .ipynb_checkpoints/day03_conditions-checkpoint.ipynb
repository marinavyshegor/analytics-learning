# День 3: Условия if/elif/else
## Практика на SEO-данных: классификация позиций ключевых слов

```python
position = 8

if position <= 3:
    print("Топ-3")
elif position <= 10:
    print("Топ-10")
elif position <= 30:
    print("Топ-30")
else:
    print("Ниже топ-30")
```

```python
position = 50

if position <= 3:
    print("Топ-3")
elif position <= 10:
    print("Топ-10")
elif position <= 30:
    print("Топ-30")
else:
    print("Ниже топ-30")
```

```python
position = int(input("Введите позицию ключевого слова: "))

if position <= 3:
    print("Топ-3")
elif position <= 10:
    print("Топ-10")
elif position <= 30:
    print("Топ-30")
else:
    print("Ниже топ-30")
```

## Выводы дня

- if — первое условие, обязательное
- elif — следующие условия, может быть много
- else — если ничего не подошло, необязательное
- Отступы важны — Python определяет блок кода по отступу
- input() всегда возвращает строку — нужно оборачивать в int()

```python
a = int(input())
print(a > 0)
```

```python
a = int(input())
print(a > 0)
```

```python
a = int(input())
print(a >= 10 and a < 100)
```
