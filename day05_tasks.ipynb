```python
a = float(input())
b = float(input())
h = input()
if h == '+':
    print(a + b)
elif h == '-':
    print(a - b)
elif h == '*':
    print(a * b)
elif h == '/':
    if b == 0:
        print('Деление на 0!')
    else:
        print(a / b)
elif h == 'mod':
    if b == 0:
        print('Деление на 0!')   
    else:
        print(a % b)
elif h == 'pow':
    print(a ** b)
elif h == 'div':
    if b == 0:
        print('Деление на 0!')
    else:
        print(a // b)
```
